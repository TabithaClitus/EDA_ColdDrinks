```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("regression_datasets_large.xlsx - Cold_Drink_Sales.csv")
X = df[["Temperature_C"]]   
y = df["Sales"]             
model = LinearRegression()
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression Equation: Sales = {slope:.2f} * Temperature + {intercept:.2f}")
sales40 = model.predict([[40]])[0]
print(f"Predicted Sales: {sales40:.2f}")
```
